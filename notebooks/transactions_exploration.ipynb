```python
import pandas as pd

# Fusionner df_trans avec df_prod sur 'product_id'
df_merged = df_trans.merge(df_prod, on='product_id', how='left')

# Afficher les premières lignes du DataFrame fusionné
print(df_merged.head())
```

```python
df_merged.columns
```

```python

# Marque préférée du client (la plus achetée en quantité)
favorite_brand = df_merged.groupby(['client_id', 'brand_name'])['quantity'].sum().reset_index()
favorite_brand = favorite_brand.loc[favorite_brand.groupby('client_id')['quantity'].idxmax(), ['client_id', 'brand_name']]

# Nombre de commandes différentes par client
num_orders = df_merged.groupby('client_id')['order_id'].nunique().reset_index().rename(columns={'order_id': 'num_orders'})

# Nombre de produits différents commandés par client
num_products = df_merged.groupby('client_id')['product_id'].nunique().reset_index().rename(columns={'product_id': 'num_products'})

# Produit acheté en plus grande quantité par client
top_quantity_product = df_merged.groupby(['client_id', 'product_id'])['quantity'].sum().reset_index()
top_quantity_product = top_quantity_product.loc[top_quantity_product.groupby('client_id')['quantity'].idxmax(), ['client_id', 'product_id']]

# Produit acheté le plus souvent par client (fréquence d'achat)
top_frequency_product = df_merged.groupby(['client_id', 'product_id'])['order_id'].nunique().reset_index()
top_frequency_product = top_frequency_product.loc[top_frequency_product.groupby('client_id')['order_id'].idxmax(), ['client_id', 'product_id']]
top_frequency_products = df_merged.groupby(['client_id', 'product_id'])['order_id'].nunique().reset_index()
top_frequency_products = top_frequency_products.sort_values(by=['client_id', 'order_id'], ascending=[True, False])

# Extraire les 2ème et 3ème produits les plus achetés
top_2_product = top_frequency_products.groupby('client_id').nth(1).reset_index()[['client_id', 'product_id']].rename(columns={'product_id': 'second_top_product'})
top_3_product = top_frequency_products.groupby('client_id').nth(2).reset_index()[['client_id', 'product_id']].rename(columns={'product_id': 'third_top_product'})

# Canal le plus utilisé par client
favorite_channel = df_merged.groupby(['client_id', 'channel'])['order_id'].count().reset_index()
favorite_channel = favorite_channel.loc[favorite_channel.groupby('client_id')['order_id'].idxmax(), ['client_id', 'channel']]

# Nombre de canaux différents utilisés par un client
num_channels = df_merged.groupby('client_id')['channel'].nunique().reset_index().rename(columns={'channel': 'num_channels'})

# Quantité moyenne par commande pour chaque client
avg_quantity_per_order = df_merged.groupby('client_id')['quantity'].mean().reset_index().rename(columns={'quantity': 'avg_quantity_per_order'})

# Catégorie produit 1 / 2 / 3 / 4 préférée du client (la plus achetée en quantité)
def favorite_category(level):
    cat = df_merged.groupby(['client_id', level])['quantity'].sum().reset_index()
    return cat.loc[cat.groupby('client_id')['quantity'].idxmax(), ['client_id', level]].rename(columns={level: f'favorite_{level}'})

favorite_cat_1 = favorite_category('prd_category_1')
favorite_cat_2 = favorite_category('prd_category_2')
favorite_cat_3 = favorite_category('prd_category_3')
favorite_cat_4 = favorite_category('prd_category_4')

# Fusionner toutes les métriques en un DataFrame final
df_client_metrics = num_orders\
    .merge(num_products, on='client_id')\
    .merge(top_quantity_product, on='client_id')\
    .merge(top_frequency_product, on='client_id')\
    .merge(favorite_channel, on='client_id')\
    .merge(num_channels, on='client_id')\
    .merge(avg_quantity_per_order, on='client_id')\
    .merge(favorite_brand, on='client_id')\
    .merge(favorite_cat_1, on='client_id')\
    .merge(favorite_cat_2, on='client_id')\
    .merge(favorite_cat_3, on='client_id')\
    .merge(favorite_cat_4, on='client_id')\
    .merge(top_2_product, on='client_id', how='left')\
    .merge(top_3_product, on='client_id', how='left')

# Afficher les premières lignes du DataFrame résultant
print(df_client_metrics.head())
```

```python
df_client_metrics.to_csv('client_metrics_Léa.csv', index=False, encoding='utf-8')
df_client_metrics.to_excel('client_metrics_Léa.xlsx', index=False, engine='openpyxl')
```
